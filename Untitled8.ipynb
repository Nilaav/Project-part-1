```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("accounting_teaching_path_dataset_4000.csv")

# Drop column A (student_id)
df = df.iloc[:, 1:14]   # B to N

df.head()
```

```python
print(df.columns)
print(df.info())
```

```python
num_cols = df.select_dtypes(include=np.number).columns

# Stats
print(df[num_cols].describe())

# Histograms
df[num_cols].hist(figsize=(12,10))
plt.tight_layout()
plt.show()
```

```python
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    print(f"\n{col} value counts:")
    print(df[col].value_counts())

    sns.countplot(x=col, data=df)
    plt.xticks(rotation=45)
    plt.title(f"{col} distribution")
    plt.show()
```

```python
from sklearn.preprocessing import OrdinalEncoder

df['difficulty_level'] = OrdinalEncoder(
    categories=[['Easy','Moderate','Hard']]
).fit_transform(df[['difficulty_level']])
```

```python
for col in num_cols:
    plt.figure()
    sns.boxplot(x='difficulty_level', y=col, data=df)
    plt.title(f"{col} vs difficulty_level")
    plt.show()
```

```python
for col in cat_cols:
    if col != 'difficulty_level':
        plt.figure()
        sns.countplot(x=col, hue='difficulty_level', data=df)
        plt.xticks(rotation=45)
        plt.title(f"{col} vs difficulty_level")
        plt.show()
```

```python
corr = df.corr(numeric_only=True)

target_corr = corr['difficulty_level'].sort_values()

print(target_corr)
```

```python
from scipy.stats import f_oneway

anova_results = {}

for col in num_cols:
    groups = [
        df[df['difficulty_level']==lvl][col]
        for lvl in df['difficulty_level'].unique()
    ]

    stat, p = f_oneway(*groups)
    anova_results[col] = p

anova_df = pd.DataFrame.from_dict(anova_results, orient='index', columns=['p_value'])

print(anova_df.sort_values('p_value'))
```

```python
low_corr = target_corr[abs(target_corr) < 0.05].index
high_p = anova_df[anova_df['p_value'] > 0.05].index

drop_cols = set(low_corr).intersection(set(high_p))

print("Columns to drop:\n", drop_cols)
```
